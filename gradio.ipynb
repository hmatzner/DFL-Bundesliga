```
MODEL_PATH = '/content/drive/MyDrive/ITC_Bundesliga/Models/model_ms4_2'
```

```
LABELS = ['challenge', 'play', 'throwin']
```

```
model = keras.models.load_model(MODEL_PATH)
```

```
def classify_gray_image(inp):
    inp = inp.reshape((-1, 224, 224, 3))
    predictions = model.predict(inp).flatten()
    confidences = {k: float(v) for k, v in zip(LABELS, predictions)}
    return confidences
```

```
import gradio as gr

gr.Interface(fn=classify_gray_image, 
             inputs=gr.Image(),
             outputs=gr.Label(),
             examples=examples).launch(share=True)
```
